
## 🧹 ## 🧹 Data Cleaning and Standardization

Performed essential data cleaning steps to prepare the dataset for analysis:
- **Date Conversion**: Converted `order_datetime` to datetime format and forward-filled missing values.
- **Numeric Conversion**: Converted `customer_support_rating` and `customer_age` to numeric types, handling invalid entries.
- **Standardization**: Cleaned text columns by stripping spaces and standardizing capitalization.
- **Placeholder Handling**: Replaced invalid entries like `'Unknown'`, `'N/A'`, `'Invalid'` with `NaN`.
- **Outlier Treatment**: Corrected unrealistic ages in `customer_age` based on median.
- **Packaging Cost**: Handled invalid entries and filled missing values with the median.
- **Weather & Ratings**: Forward-filled or imputed missing values where appropriate.

These steps ensure a clean and consistent dataset for further analysis.



```python
# Convert 'order_datetime' to datetime and handle errors
df['order_datetime'] = pd.to_datetime(df['order_datetime'], errors='coerce')
df['order_datetime'] = df['order_datetime'].fillna(method="ffill")

# Clean string columns: remove spaces and standardize capitalization
df['customer_name'] = df['customer_name'].str.strip().str.title()
df['platform'] = df['platform'].str.strip().str.title()
df['category'] = df['category'].str.strip().str.title()
df['weather'] = df['weather'].str.strip().str.title()
df['city'] = df['city'].str.strip().str.title()

# Replace unwanted text entries with NaN
df.replace([" ", 'Unknown', 'N/A', 'Invalid'], np.nan, inplace=True)

# Correct invalid 'customer_age' entries and convert to numeric
df['customer_age'] = df['customer_age'].replace('thirty', 30)
df['customer_age'] = pd.to_numeric(df['customer_age'], errors='coerce')
median_age = df['customer_age'].median()
df.loc[(df['customer_age'] < 10) | (df['customer_age'] > 90), 'customer_age'] = median_age

# Clean and impute missing values in 'packaging_cost'
df['packaging_cost'] = df['packaging_cost'].replace(['free', '', 'na', 'n/a', 'unknown'], np.nan)
df['packaging_cost'] = pd.to_numeric(df['packaging_cost'], errors='coerce')
df['packaging_cost'].fillna(df['packaging_cost'].median(), inplace=True)


# Forward fill missing 'weather' values
df['weather'] = df['weather'].fillna(method='ffill')

# Check remaining missing values
df.isna().sum()
```


## 🛠️ Feature Engineering & Data Preparation

We perform key feature engineering steps to calculate metrics like Revenue, Discounts, Net Revenue, Fulfillment Cost, and Profit/Loss.  
Time-based features like Order Hour and Day are extracted for analysis.  
After that, columns are renamed, reordered, and key monetary fields are formatted with ₹ for clarity.

> This prepares the data for analysis while preserving the original dataset.



```python
# ✅ Step 1 — Make a Working Copy 
df_work = df.copy()

# ✅ Step 2 — Feature Engineering on Working Copy (using lowercase column names)
df_work['Revenue'] = df_work['product_MRP'] * df_work['quantity']
df_work['Discounted_Price'] = df_work['product_MRP'] * (1 - df_work['discount_percent'] / 100)
df_work['Discounted_Amount'] = df_work['Revenue'] * (df_work['discount_percent'] / 100)

df_work['NetRevenue'] = df_work['Revenue'] - df_work['Discounted_Amount']
df_work['Total_fulfillment_cost'] = (
    (df_work['quantity'] * df_work['product_cost_price']) +
    df_work['packaging_cost'] + df_work['operational_cost'] + df_work['delivery_cost']
)

df_work['Profit/Loss'] = df_work['NetRevenue'] - df_work['Total_fulfillment_cost']

df_work['Order_Hour'] = df_work['order_datetime'].dt.hour
df_work['Order_Hour'] = df_work['Order_Hour'].map(lambda x: (
    'Morning' if 5 <= x < 12 else
    'Afternoon' if 12 <= x < 17 else
    'Evening' if 17 <= x < 21 else
    'Night'
))
df_work['Order_Day'] = df_work['order_datetime'].dt.day_name()

df_work['Delivery_Delay_Minutes'] = df_work['delivery_time_minutes'] - df_work['estimated_delivery']

# ✅ Step 3 — Rename Columns AFTER Feature Engineering
df_work.columns = df_work.columns.str.title()

# ✅ Step 4 — Reorder Columns Using Title Case
new_column_order = [
    'Order_Id', 'Order_Datetime', 'Order_Hour', 'Order_Day',
    'Customer_Id', 'Customer_Name', 'Customer_Age', 'Gender', 'Platform', 'City',
    'Category', 'Product_Name', 'Product_Mrp', 'Discount_Percent', 'Discounted_Amount',
    'Discounted_Price', 'Quantity', 'Product_Cost_Price',
    'Packaging_Cost', 'Operational_Cost', 'Delivery_Cost', 'Total_Fulfillment_Cost',
    'Dark_Store_Location', 'Delivery_Partner_Type', 'Delivery_Distance_Km',
    'Estimated_Delivery', 'Delivery_Time_Minutes', 'Delivery_Delay_Minutes',
    'Delivery_Rating', 'Customer_Support_Rating', 'Weather', 'Returned/Cancel',
    'Revenue', 'Netrevenue', 'Profit/Loss'
]

df_work = df_work[[col for col in new_column_order if col in df_work.columns]]

# ✅ Step 5 — Format Monetary Columns with Rupee Symbol
def format_rupee(x):
    return f"₹{x:,.0f}"

df_formatted = df_work.copy()

for col in [
    'Product_Mrp', 'Product_Cost_Price', 'Discounted_Amount', 'Discounted_Price',
    'Packaging_Cost', 'Delivery_Cost', 'Operational_Cost', 'Revenue',
    'Netrevenue', 'Profit/Loss', 'Total_Fulfillment_Cost'
]:
    if col in df_formatted.columns:
        df_formatted[col] = df_formatted[col].apply(format_rupee)

# ✅ Step 6 — Preview the Formatted DataFrame
df_formatted.head()
```

## Q1) Which product categories generate the highest revenue and order volumes?

**Business context:**  
Identifying top-selling categories helps optimize assortment, pricing, and marketing efforts by focusing on the biggest revenue drivers.



```python

category_summary = (
    df_work.groupby('Category', as_index=False)
    .agg(Order_Count=('Order_Id', 'count'), Revenue=('Revenue', 'sum'))
    .sort_values('Revenue', ascending=False)
)

# ✅ Plot
plt.figure(figsize=(18, 7))
bars = sns.barplot(
    data=category_summary,
    x='Revenue',
    y='Category',
    color='#4575b4'
)

# ✅ Highlight top bar
bars.patches[0].set_facecolor('#f46d43')

# ✅ Add labels
for p in bars.patches:
    plt.text(
        p.get_width() * 1.01,
        p.get_y() + p.get_height() / 2,
        f'₹{p.get_width():,.0f}',
        ha='left', va='center', fontsize=12, fontweight='bold'
    )

plt.title('Total Revenue by Product Category', fontsize=20, fontweight='bold')
plt.xlabel('Revenue (INR)', fontsize=16, fontweight='bold')
plt.ylabel('Product Category', fontsize=16, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'₹{int(x):,}'))
plt.tight_layout()
plt.show()
```

### Insights
- **Groceries contribute the highest revenue** in the Q-commerce platform, followed by **Personal Care and Beverages**, making them the top-performing categories.
- **Frozen Food and Household items generate the least revenue**, suggesting potential areas for growth or reevaluation of inventory strategy.


## Q2) What times of day see the most orders placed, and how does this affect delivery delays?

**Business context:**  
Understanding peak order hours and their impact on delivery performance enables better staffing, routing, and customer experience during rush times.


```python

# Ordered Categories for Consistent Plotting
time_order = ['Morning', 'Afternoon', 'Evening', 'Night']
df_work['Time_Of_Day'] = pd.Categorical(df_work['Order_Hour'], categories=time_order, ordered=True)

# Aggregate Data
time_orders = df_work.groupby('Time_Of_Day', as_index=False)['Order_Id'].count()
time_delays = df_work.groupby('Time_Of_Day', as_index=False)['Delivery_Delay_Minutes'].mean() 

# ✅ Proceed only if valid data exists
if not time_orders.empty and time_orders['Order_Id'].max() > 0:
    fig, ax1 = plt.subplots(figsize=(14, 7))

    # ✅ Barplot for Order Count by Time of Day
    sns.barplot(data=time_orders, x='Time_Of_Day', y='Order_Id',
                color='#4575b4', ax=ax1, alpha=0.9)

    ax1.set_title('Orders and Average Delivery Delays by Time of Day', pad=15)
    ax1.set_xlabel('Time of Day')
    ax1.set_ylabel('Number of Orders')
    ax1.grid(axis='y', linestyle='--', alpha=0.3)

    # ✅ Annotate Order Counts on Top of Bars
    for p in ax1.patches:
        ax1.text(
            p.get_x() + p.get_width() / 2,
            p.get_height() + max(time_orders['Order_Id']) * 0.015,
            f'{int(p.get_height())}',
            ha='center', va='bottom',
            fontsize=12, fontweight='bold'
        )

    # ✅ Lineplot on Secondary Y-Axis for Average Delivery Delay
    ax2 = ax1.twinx()
    sns.lineplot(data=time_delays, x='Time_Of_Day', y='Delivery_Delay_Minutes',
                 color='#f46d43', marker='o', markersize=9, linewidth=2.5, ax=ax2, label='Average Delay')

    ax2.set_ylabel('Avg Delivery Delay (min)', color='#f46d43')
    ax2.tick_params(axis='y', colors='#f46d43')
    ax2.grid(False)  # Disable secondary grid for clarity

    # ✅ Annotate Delay Points on Line
    for i, delay in enumerate(time_delays['Delivery_Delay_Minutes']):
        ax2.text(
            i,
            delay + max(time_delays['Delivery_Delay_Minutes']) * 0.03,
            f"{delay:.1f}",
            color='white',
            ha='center', va='bottom',
            fontsize=12, fontweight='bold'
        )

    # ✅ Adjust Secondary Y-Axis Range for Better Visual Spacing
    ax2.set_ylim(
        max(time_delays['Delivery_Delay_Minutes'].min() * 0.85, 1),
        time_delays['Delivery_Delay_Minutes'].max() * 1.15
    )

    ax2.legend(loc='upper left', fontsize=12, frameon=False)

    plt.tight_layout()
    plt.show()
else:
    print("No valid data available for plotting.") 
```

### Insights
- **Evening and night time slots have the highest number of orders**, indicating peak customer activity during these hours.
- **Average delivery delay remains highest in the Evening (2.5 min) and improves towards the Night (2.0 min)**, suggesting better delivery efficiency during late hours despite high order volumes.


## Q3) What percentage of orders miss promised delivery times, and what patterns exist in these late deliveries?

**Business context:**  
On-time delivery is a key differentiator for quick commerce platforms; analyzing delays uncovers operational bottlenecks affecting customer satisfaction.


```python
# ✅ Step 1 — Donut Chart Data Preparation
df_work['Is_Late'] = df_work['Delivery_Delay_Minutes'] > 0
late_counts = df_work['Is_Late'].value_counts().values
donut_colors = ['#4575b4', '#f46d43']
labels = ['On Time Deliveries', 'Late Deliveries']

# ✅ Step 2 — Heatmap Data Preparation
late_df = df_work[df_work['Is_Late']].copy()

# Ensure correct order of categories for better heatmap layout
late_df['Order_Day'] = pd.Categorical(late_df['Order_Day'], 
                                      categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 
                                      ordered=True)
late_df['Weather'] = pd.Categorical(late_df['Weather'], 
                                     categories=['Sunny', 'Clear', 'Cloudy', 'Foggy', 'Rainy', 'Stormy'], 
                                     ordered=True)

# Group & calculate percentage of late orders per weather and day
late_by_day_weather = late_df.groupby(['Order_Day', 'Weather'], observed=True).size().reset_index(name='Late_Count')
late_by_day_weather['Percentage'] = late_by_day_weather.groupby('Order_Day')['Late_Count'].transform(lambda x: x / x.sum())
heatmap_data = late_by_day_weather.pivot(index='Weather', columns='Order_Day', values='Percentage')

# ✅ Step 3 — Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8), gridspec_kw={'width_ratios': [1, 1.6]})

# 🎯 Donut Chart — On Time vs Late Deliveries
wedges, texts, autotexts = ax1.pie(
    late_counts, 
    autopct=lambda pct: f"{pct:.1f}%\n({int(pct/100*sum(late_counts)):,})",
    startangle=90, colors=donut_colors, explode=(0.04, 0.04), pctdistance=0.8,
    wedgeprops=dict(width=0.35, edgecolor='w')
)
# Style donut text & add center circle
for text in texts + autotexts:
    text.set_fontsize(14)
    text.set_fontweight('bold')
ax1.add_artist(plt.Circle((0, 0), 0.55, fc='white'))
ax1.set_title('On-Time vs. Late Deliveries', fontsize=18, fontweight='bold')
ax1.legend(wedges, labels, title="Delivery Status", loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2, fontsize=12)

# 📊 Heatmap — Late Delivery Pattern by Day & Weather
sns.heatmap(heatmap_data, annot=True, fmt=".1%", cmap='RdBu_r', linewidths=0.5, 
            linecolor='white', vmin=0, vmax=0.3, annot_kws={"fontsize": 11, "fontweight": "bold"}, 
            ax=ax2, cbar_kws={'label': 'Late Delivery %'})

ax2.set_title('Pattern of Late Deliveries by Day of Week and Weather', fontsize=18, fontweight='bold', pad=15)
ax2.set_xlabel('Day of Week', fontsize=16, fontweight='bold')
ax2.set_ylabel('Weather', fontsize=16, fontweight='bold')
ax2.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.subplots_adjust(bottom=0.1)
plt.show()
```

### Insights
- **Around 34% of orders in Q-commerce are delivered late**, indicating a need for improved delivery efficiency and logistics planning.
- **Late deliveries are more frequent during rainy and Stormy Weather**, especially on **Weekends**, suggesting that weather conditions and weekend demand affect delivery performance.
 

## Q4) How do weather conditions impact order volumes and delivery delays?

**Business context:**  
Adverse weather often disrupts deliveries and changes demand patterns. Insights here guide staffing, delivery promises, and inventory planning.


```python
# ✅ Ensure consistent weather order in df_work
weather_order = ['Sunny', 'Clear', 'Cloudy', 'Foggy', 'Rainy', 'Stormy']
df_work['Weather'] = pd.Categorical(df_work['Weather'], categories=weather_order, ordered=True)

# ✅ 1) Orders by Weather
orders_by_weather = df_work.groupby('Weather', as_index=False).agg(Order_Count=('Order_Id', 'count'))

# ✅ 2) Average Delay by Weather
avg_delay_by_weather = df_work.groupby('Weather', as_index=False).agg(Avg_Delay=('Delivery_Delay_Minutes', 'mean'))
 
primary_color = '#4575b4'
highlight_color = '#f46d43'

fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# ✅ Left Plot — Orders by Weather
bars1 = sns.barplot(data=orders_by_weather, x='Order_Count', y='Weather', color=primary_color, order=weather_order, ax=axes[0])
bars1.patches[orders_by_weather['Order_Count'].idxmax()].set_facecolor(highlight_color)

axes[0].set_title('Orders by Weather Condition',fontsize=20,fontweight='bold')
axes[0].set_xlabel('Order Count',fontsize=17,fontweight='bold')
axes[0].set_ylabel('Weather',fontsize=20,fontweight='bold')
axes[0].grid(axis='x', linestyle='--', alpha=0.3)

for p in bars1.patches:
    axes[0].text(p.get_width() + max(orders_by_weather['Order_Count']) * 0.01,
                 p.get_y() + p.get_height() / 2,
                 f'{int(p.get_width())}', va='center', fontsize=12, fontweight='bold')

# ✅ Right Plot — Avg Delivery Delay by Weather
bars2 = sns.barplot(data=avg_delay_by_weather, x='Avg_Delay', y='Weather', color=primary_color, order=weather_order, ax=axes[1])
bars2.patches[avg_delay_by_weather['Avg_Delay'].idxmax()].set_facecolor(highlight_color)

axes[1].set_title('Average Delivery Delay by Weather',fontsize=20,fontweight='bold')
axes[1].set_xlabel('Avg Delay (min)',fontsize=17,fontweight='bold')
axes[1].grid(axis='x', linestyle='--', alpha=0.3)

for p in bars2.patches:
    axes[1].text(p.get_width() + max(avg_delay_by_weather['Avg_Delay']) * 0.01,
                 p.get_y() + p.get_height() / 2,
                 f'{p.get_width():.1f}', va='center', fontsize=12, fontweight='bold')

plt.suptitle('Order Volume and Delivery Delays by Weather Condition', fontsize=18, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

### Insights
- **Clear and Sunny conditions see the highest order volumes**, reflecting customer preference to order in favorable weather.
- **Rainy weather causes the highest average delivery delays (3.1 minutes)**, followed by Stormy (2.2 mins), highlighting the logistical challenges faced during poor weather.


## Q5) What product categories and discount strategies drive higher average order value in Q-commerce 

**Business context:**  
Analyzing AOV across demographics reveals which segments contribute more revenue per order, informing personalized marketing and promotional strategies.

---

```python
from matplotlib.ticker import FuncFormatter
primary_color = '#4575b4'
highlight_color = '#fb7c5c'

# ✅ AOV by Category
category_orders = df_work.groupby('Category', as_index=False).agg(Order_Id=('Order_Id', 'nunique'), Revenue=('Revenue', 'sum'))
category_orders['AOV'] = category_orders['Revenue'] / category_orders['Order_Id']
category_orders = category_orders.sort_values('AOV')

# ✅ AOV by Discount Level
df_work['Discount_Level'] = pd.cut(df_work['Discount_Percent'], bins=[-0.01, 10, 25, 100], labels=['Low (0-10%)', 'Medium (10-25%)', 'High (25%+)'])
discount_aov = df_work.groupby('Discount_Level', as_index=False).agg(Order_Id=('Order_Id', 'nunique'), Revenue=('Revenue', 'sum'))
discount_aov['AOV'] = discount_aov['Revenue'] / discount_aov['Order_Id']
discount_aov = discount_aov.sort_values('AOV', ascending=False)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 8))

# ✅ Plot 1 — AOV by Category
bars1 = sns.barplot(data=category_orders, x='AOV', y='Category', color=primary_color, ax=ax1, edgecolor='black')
bars1.patches[list(category_orders.index).index(category_orders['AOV'].idxmax())].set_facecolor(highlight_color)
ax1.set_title('Average Order Value by Product Category', fontsize=22, fontweight='bold')
ax1.set_xlabel('Average Order Value (INR)', fontsize=20, fontweight='bold')
ax1.set_ylabel('Product Category', fontsize=20, fontweight='bold')
ax1.grid(axis='x', linestyle='--', alpha=0.3)
ax1.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

for p in bars1.patches:
    ax1.text(p.get_width() + max(category_orders['AOV']) * 0.02,
             p.get_y() + p.get_height() / 2,
             f'{p.get_width():,.0f}', ha='left', va='center', fontsize=12, fontweight='bold')

# ✅ Plot 2 — AOV by Discount Level (Highlight Medium)
bars2 = sns.barplot(data=discount_aov, x='Discount_Level', y='AOV', color=primary_color, ax=ax2, edgecolor='black')
highlight_idx = discount_aov[discount_aov['Discount_Level'] == 'Medium (10-25%)'].index[0]
bars2.patches[highlight_idx].set_facecolor(highlight_color)
ax2.set_title('Average Order Value by Discount Level', fontsize=22, fontweight='bold')
ax2.set_xlabel('Discount Level', fontsize=18, fontweight='bold')
ax2.set_ylabel('Average Order Value (INR)', fontsize=18, fontweight='bold')
ax2.grid(axis='y', linestyle='--', alpha=0.3)
ax2.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))
axes[0].set_xticklabels(axes[0].get_xticklabels(), fontsize=18)
axes[0].set_yticklabels(axes[0].get_yticklabels(), fontsize=18)

axes[1].set_xticklabels(axes[1].get_xticklabels(), fontsize=18)
axes[1].set_yticklabels(axes[1].get_yticklabels(), fontsize=18)


for p in bars2.patches:
    ax2.text(p.get_x() + p.get_width() / 2,
             p.get_height() + max(discount_aov['AOV']) * 0.025,
             f'{p.get_height():,.0f}', ha='center', va='bottom', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Insights
- **Groceries and Personal Care** lead with the highest average order values, indicating higher consumer spending in these categories.
- Orders with **medium-level discounts (10–25%) yield the highest average value (₹308)**, suggesting an optimal discount band for maximizing basket size without excessive revenue loss.


## Q6) How do discounts and promotions affect net profit per order?

**Business context:**  
Excessive discounting can drive sales but erode margins. Understanding the relationship between discount levels and profitability helps create sustainable promotion strategies.


```python

import matplotlib.ticker as mtick
primary_color = '#4575b4'
highlight_color = '#fb7c5c'  # ✅ Slightly lighter coral

# ✅ 1) Discount Bins
discount_bins = pd.cut(
    df_work['Discounted_Amount'],
    bins=[0, 20, 50, 100, df_work['Discounted_Amount'].max()],
    labels=['Low', 'Medium', 'High', 'Very High'],
    ordered=True
)
df_work['Discount_Level'] = discount_bins

# ✅ 2) Aggregate
profit_by_discount = (
    df_work.groupby('Discount_Level', observed=True)
    .agg(
        Avg_Net_Profit=('Profit/Loss', 'mean'),
        Order_Count=('Order_Id', 'nunique')
    )
    .reset_index()
)

# ✅ 3) Plot
plt.figure(figsize=(12, 6))
bars = sns.barplot(
    data=profit_by_discount,
    x='Discount_Level',
    y='Avg_Net_Profit',
    color=primary_color
)

# ✅ 4) Highlight the bar with lowest Avg_Net_Profit
min_idx = profit_by_discount['Avg_Net_Profit'].idxmin()
bars.patches[min_idx].set_facecolor(highlight_color)

# ✅ 5) Zero Line
plt.axhline(0, color='black', linewidth=1.2, linestyle='--')

# ✅ 6) Annotate Bars
for bar, count, profit in zip(bars.patches, profit_by_discount['Order_Count'], profit_by_discount['Avg_Net_Profit']):
    bar_x = bar.get_x() + bar.get_width() / 2
    label = f"₹{profit:,.0f}\n({count} orders)"
    text_color = 'white' if profit >= 0 else 'black'
    plt.text(
        bar_x,
        profit * 0.5,
        label,
        ha='center', va='center',
        fontsize=11, fontweight='bold', color=text_color
    )

# ✅ 7) Format Y-axis
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('₹{x:,.0f}'))

# ✅ 8) Titles and Layout
plt.title('Average Net Profit/Loss by Discount Level', pad=20)
plt.xlabel('Discount Level')
plt.ylabel('Average Net Profit per Order (INR)')
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()

```

### Insights
- **As discount levels increase, the average net profit per order declines sharply** — indicating that deeper discounts significantly reduce profitability.
- **Very high discounts are the least frequent** , making up the smallest portion of orders, suggesting such deep discounting is used sparingly but is the most unprofitable.

## 🧐 Key Business Question: How Can Quick Commerce Achieve Profitability in India?

Despite explosive growth, India’s quick commerce players remain deeply unprofitable. In FY24 alone, Zepto posted losses of ₹1,249 Cr, while Swiggy Instamart widened its annual loss to approximately ₹1,887 Cr. Blinkit (Zomato's Q-commerce arm) also reported a ₹1,030 Cr core loss in Q3 FY25. These figures underscore a critical challenge — high customer acquisition costs, deep discounting, and operational inefficiencies continue to erode margins at scale.

To move toward sustainable profitability, Q-commerce platforms must focus on **increasing Average Order Value (AOV)** through strategies like bundling, minimum order thresholds, and upselling. Introducing **high-margin product categories** (e.g., personal care, electronics accessories) and developing **in-house or private-label brands** can significantly boost profit per order, as they offer greater control over pricing and sourcing.

Additionally, strategically expanding **dark stores in high-demand areas** can reduce delivery times and improve fulfillment efficiency — a pattern confirmed in our analysis. Paired with smarter discounting, route optimization, and weather-adaptive delivery strategies, these changes can help shift the industry’s focus from revenue growth to long-term viability and margin improvement.


---

## ✅ Recommendations for Q-Commerce Business
- **Increase Average Order Value (AOV)**  
  Encourage larger baskets through bundling, cross-selling, and minimum order thresholds to improve profit per delivery.

- **Introduce High-Margin and In-House Products**  
  Expand into high-margin categories like personal care and electronics accessories. Launch and promote in-house (private label) products to gain control over sourcing and pricing, improving gross margins.

- **Optimize Delivery Logistics**  
  Improve last-mile delivery through route optimization and better scheduling, especially during peak hours and adverse weather conditions to reduce fulfillment costs.

- **Expand Dark Store Network Strategically**  
  Invest in expanding dark stores in high-demand zones. As shown in the analysis, shorter delivery distances are correlated with faster delivery times and improved customer experience, which in turn supports higher retention and revenue growth.

- **Refine Discount Strategy**  
  Reduce deep discounting on low-margin categories and redirect promotional efforts toward products that offer higher profitability.

- **Focus on High-Value Customer Segments**  
  Retarget and retain the 25–34 age group, which represents the most active and high-value customer segment. Consider personalized offers for other underrepresented age groups to broaden reach.

- **Weather-Based Contingency Planning**  
  Use historical weather-delay patterns to plan delivery fleet allocation and improve reliability during rainy or stormy conditions.

- **Monitor Basket Sizes and Fulfillment Costs**  
  Avoid profitability loss from large, low-margin baskets by introducing packaging/delivery fees or intelligent packing strategies to optimize fulfillment expenses.
